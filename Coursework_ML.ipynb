# 1. Описание данных

```python
### YOUR CODE HERE ###
import pandas as pd
import numpy as np

# занрузка данных
telecom_data = pd.read_csv('telecom_users.csv')
print(telecom_data.info())
print(telecom_data.head())
```

Общая сумма является типом Object вместо float64, видимо там присутствуют неправильные значения

```python
# замена нечислового значения на медианное значение

telecom_data['TotalCharges'] = pd.to_numeric(telecom_data['TotalCharges'], errors='coerce')
telecom_data['TotalCharges'] = telecom_data['TotalCharges'].fillna((telecom_data['TotalCharges'].median()))

print(telecom_data.info())
```

```python
# валидация данных

features = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn']
for column in features:
    print('-------------------------')
    print(telecom_data[column].value_counts())
```

```python
# изменение типов на категориальные
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

telecom_data['gender'] = telecom_data['gender'].replace(to_replace=['Male', 'Female'], value=[0, 1])
telecom_data['Partner'] = telecom_data['Partner'].replace(to_replace=['No', 'Yes'], value=[0, 1])
telecom_data['Dependents'] = telecom_data['Dependents'].replace(to_replace=['No', 'Yes'], value=[0, 1])
telecom_data['PhoneService'] = telecom_data['PhoneService'].replace(to_replace=['No', 'Yes'], value=[0, 1])
telecom_data['MultipleLines'] = telecom_data['MultipleLines'].replace(to_replace=['No', 'Yes', 'No phone service'], value=[0, 1, 2])
telecom_data['InternetService'] = telecom_data['InternetService'].replace(to_replace=['No', 'Fiber optic', 'DSL'], value=[0, 1, 2])
telecom_data['OnlineSecurity'] = telecom_data['OnlineSecurity'].replace(to_replace=['No', 'Yes', 'No internet service'], value=[0, 1, 2])
telecom_data['OnlineBackup'] = telecom_data['OnlineBackup'].replace(to_replace=['No', 'Yes', 'No internet service'], value=[0, 1, 2])
telecom_data['DeviceProtection'] = telecom_data['DeviceProtection'].replace(to_replace=['No', 'Yes', 'No internet service'], value=[0, 1, 2])
telecom_data['TechSupport'] = telecom_data['TechSupport'].replace(to_replace=['No', 'Yes', 'No internet service'], value=[0, 1, 2])
telecom_data['StreamingTV'] = telecom_data['StreamingTV'].replace(to_replace=['No', 'Yes', 'No internet service'], value=[0, 1, 2])
telecom_data['StreamingMovies'] = telecom_data['StreamingMovies'].replace(to_replace=['No', 'Yes', 'No internet service'], value=[0, 1, 2])
telecom_data['Contract'] = telecom_data['Contract'].replace(to_replace=['Month-to-month', 'Two year', 'One year'], value=[0, 1, 2])
telecom_data['PaperlessBilling'] = telecom_data['PaperlessBilling'].replace(to_replace=['No', 'Yes'], value=[0, 1])
telecom_data['PaymentMethod'] = telecom_data['PaymentMethod'].replace(to_replace=['Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic)'], value=[0, 1, 2, 3])
telecom_data['Churn'] = telecom_data['Churn'].replace(to_replace=['No', 'Yes'], value=[0, 1])

#for column in features:
#    telecom_data[column] = label_encoder.fit_transform(telecom_data[column])
telecom_data.info()
```

# 3. Построение моделей для прогнозирования оттока

### Так как целевая переменная принимает конечное число значений (0,1), мы имеем дело с здачей классификации

```python
### YOUR CODE HERE ###
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
#from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn import svm
```

### Деление данных

```python
X = telecom_data.drop(['Unnamed: 0', 'customerID','Churn'], axis=1)
y = telecom_data.Churn
```

### Обучение моделей

Использованы 4 базовые модели:<br>
1.KNeighborsClassifier<br>
2.SVC<br>
3.RandomForestClassifier<br>
4.LogisticRegression

```python
knc = KNeighborsClassifier(n_neighbors=18)
svc = svm.SVC(probability=True)
rfc = RandomForestClassifier(n_estimators=70, max_depth=2, random_state=0)
lr  = LogisticRegression(penalty='l2', tol=0.01, max_iter=300)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

def predict_model(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    predict = model.predict(x_test)
    score = accuracy_score(y_test, predict)
    print(f'Accuracy: {score}')
    return predict
```

### Кросс-валидация

```python
kfold = 10
itog_val = {}

scores = cross_val_score(knc, x_train, y_train, cv = kfold)
itog_val['KNeighborsClassifier'] = scores.mean()
scores = cross_val_score(svc, x_train, y_train, cv = kfold)
itog_val['SVC'] = scores.mean()
scores = cross_val_score(rfc, x_train, y_train, cv = kfold)
itog_val['RandomForestClassifier'] = scores.mean()
scores = cross_val_score(lr, x_train, y_train, cv = kfold)
itog_val['LogisticRegression'] = scores.mean()
```

```python
from pandas import DataFrame

DataFrame.from_dict(data = itog_val, orient='index').plot(kind='bar', legend=False)
```

### Предсказания по всем моделям с параметрами по умолчанию

```python
print('KNeighborsClassifier')
knc_pred = predict_model(knc, x_train, x_test, y_train, y_test)
#print('GaussianNB')
#gnb_pred = predict_model(gnb, x_train, x_test, y_train, y_test)
print('SVC')
svc_pred = predict_model(svc, x_train, x_test, y_train, y_test)
print('RandomForestClassifier')
rfc_pred = predict_model(rfc, x_train, x_test, y_train, y_test)
print('LogisticRegression')
lr_pred = predict_model(lr, x_train, x_test, y_train, y_test)
#regress_pred = predict_model(regress, x_train, x_test, y_train, y_test)
```

 ### Настройка гипер параметры

```python
from sklearn.model_selection import GridSearchCV

knc_params = {'n_neighbors': np.arange(1, 100, 2)} # KNeighborsClassifier
rfc_params = {'n_estimators': range(50, 300, 50), # RandomForestClassifier
              'min_samples_leaf': range(1, 10)}
svc_params = {#'kernel': ['linear', 'rbf'], # SVC
              'C': np.arange(0.1, 1, 0.2)}
c_values = np.logspace(-2, 3, 500)

skf = StratifiedKFold(n_splits=5, shuffle=True)

gsc_knc = GridSearchCV(knc, knc_params, cv=skf)

gsc_svc = GridSearchCV(svc, svc_params, cv=skf)
gsc_rfc = GridSearchCV(rfc, rfc_params, cv=skf)
gsc_lr = LogisticRegressionCV(Cs=c_values, cv=skf, verbose=1, n_jobs=-1, max_iter=100)

print('KNeighborsClassifier')
knc_pred = predict_model(gsc_knc, x_train, x_test, y_train, y_test)
print('SVC')
svc_pred = predict_model(gsc_svc, x_train, x_test, y_train, y_test)
print('RandomForestClassifier')
rfc_pred = predict_model(gsc_rfc, x_train, x_test, y_train, y_test)
print('LogisticRegression')
lr_pred  = predict_model(gsc_lr,  x_train, x_test, y_train, y_test)
```

### XGBoost

```python
import xgboost as xgb

dtrain = xgb.DMatrix(x_train, y_train)
dtest = xgb.DMatrix(x_test, y_test)

params = {
   'max_depth': 3,
   'eta': 0.3, 
   'silent': 1, 
   'objective': 'multi:softprob',
   'num_class': 3}
num_rounds = 50

xgb_model = xgb.train(params, dtrain, num_rounds)
evals = [(dtest, 'test'), (dtrain, 'train')]
xgb_model = xgb.train(params, dtrain, num_rounds, evals=evals)
preds = xgb_model.predict(dtest)
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score

best_preds = np.asarray([np.argmax(line) for line in preds])
knc_accuracy = roc_auc_score(y_test, best_preds)
print(f'XGBoost Accuracy: {knc_accuracy}')
```

### Стекинг

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

models = [gsc_knc, gsc_svc, gsc_rfc, gsc_lr]

xgb_params = {'n_estimators': range(10, 100, 5),
              'eta': np.arange(0.1, 1., .1),
              'min_child_weight': range(1, 10, 1),
              'subsample': np.arange(0.1, 1., 0.2)}

xgb = XGBClassifier(random_state=17)
gsc_xgb = GridSearchCV(xgb, xgb_params, cv=skf)

meta_mtrx = np.empty((x_test.shape[0], len(models)))
meta_mtrx[:,0] = knc_pred
meta_mtrx[:,1] = svc_pred
meta_mtrx[:,2] = rfc_pred
meta_mtrx[:,3] = lr_pred

meta_model = gsc_xgb.fit(meta_mtrx, y_test)

meta_mtrx_test = np.empty((x_test.shape[0], len(models))) 
for n, model in enumerate(models):
    meta_mtrx_test[:, n] = model.predict(x_test)

meta_predict = meta_model.predict(meta_mtrx_test)
print(f'Stacking AUC: {roc_auc_score(y_test, meta_predict)}')
```

### Стекинг с функцией

```python
import numpy as np
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score

def stacking(models, meta_alg, data_train, targets_train, data_test, targets_test=None, random_state=None, test_size=None, cv=5):
    if test_size is None:
        
        meta_mtrx = np.empty((data_train.shape[0], len(models)))
        for n, model in enumerate(models):
            meta_mtrx[:,n] = cross_val_predict(model, data_train, targets_train, cv, method='predict')
            model.fit(data_train, targets_train)
            
        meta_model = meta_alg.fit(meta_mtrx, targets_train)

        meta_mtrx_test = np.empty((data_test.shape[0], len(models))) 
        for n, model in enumerate(models):
            meta_mtrx_test[:, n] = model.predict(data_test)

        meta_predict = meta_alg.predict(meta_mtrx_test)
            
        if targets_test is not None:
            print(f'Stacking AUC: {roc_auc_score(targets_test, meta_predict)}')
            
        return meta_predict
    
    elif test_size > 0 and test_size < 1:
        train, valid, train_true, valid_true = train_test_split(data_train, targets_train, train_size=test_size, random_state=random_state)
        meta_mtrx = np.empty((valid.shape[0], len(models)))

        for n, model in enumerate(models):
            model.fit(train, train_true)
            meta_mtrx[:, n] = model.predict(valid)

        meta_model = meta_alg.fit(meta_mtrx, valid_true)

        meta_mtrx_test = np.empty((data_test.shape[0], len(models))) 
        for n, model in enumerate(models):
            meta_mtrx_test[:, n] = model.predict(data_test)

        meta_predict = meta_alg.predict(meta_mtrx_test)

        if targets_test is not None:
            print(f'Stacking AUC: {roc_auc_score(targets_test, meta_predict)}')

        return meta_predict
            
    else:
        raise ValueError("test_size must be between 0 and 1")
        
predict = stacking(models, gsc_xgb, x_train, y_train, x_test, cv=None)

print(f'Stacking AUC: {roc_auc_score(y_test, predict)}')
```
